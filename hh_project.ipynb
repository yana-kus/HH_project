Создание нового признака «Образование», который должен иметь четыре категории: «высшее», «неоконченное высшее», «среднее специальное» и «среднее». Удаление столбца  «Образование и ВУЗ».


```python
# функция -преобразование 
def get_education(arg):
    arg_list=arg.split(' ')# разбиваем по пробелу
    arg_type=arg_list[:3]# выделяем первые три слова
    arg_type=' '.join(arg_type) #объединяем в строку первые три слова
    if 'Высшее' in arg_type:
        return 'высшее'
    elif 'Неоконченное высшее' in arg_type:
        return 'неоконченное высшее'
    elif 'Среднее специальное' in arg_type:
        return 'среднее специальное'
    elif 'Среднее образование' in arg_type:
        return 'среднее' 

hh_data['Образование'] = hh_data['Образование и ВУЗ'].apply(get_education) 
hh_data['Образование'].unique() # проверка работы функции
```

Преобразование признака "Опыт работы". Создание нового признака "Опыт работы(месяц)". Удаление столбца "Опыт работы".

```python
#выделяем общий опыт работы соискателя в месяцах, создаем новый признак "опыт работы (месяц)"
def get_experience(arg):

    if arg is np.nan or arg == 'Не указано':
        return None
    year_words=['год', 'года', 'лет']
    month_words=['месяц', 'месяца', 'месяцев']
    arg_splitted = arg.split(' ')[:7]
    years = 0
    months = 0

    for index, item in enumerate(arg_splitted):
        if item in year_words:
            years = int(arg_splitted[index-1])
        if item in month_words:
            months = int(arg_splitted[index-1])
    return int(years*12 + months)

hh_data['Опыт работы (месяц)'] = hh_data['Опыт работы'].apply(get_experience)
```

3.4. Процент соискателей живущих в Санкт-Петербурге. Процент соискателей готовых одновременно и к переездам и к командировкам.

```python
# функция преобразование городов к четырем категориям. 
def city_(arg):
    city = arg.split(' ')[0]
    #print(city)
    million_cities = ['Новосибирск', 'Екатеринбург', 'Нижний Новгород', 'Казань', 'Челябинск', 'Омск', 'Самара', 'Ростов-на-Дону', 'Уфа', 'Красноярск', 'Пермь', 'Воронеж', 'Волгоград' ]
    if city in million_cities:
        
        return 'город-миллионник'
    elif city in ['Москва', 'Санкт-Петербург']:
        
        return city
    else:
        
        return 'другие'
hh_data['Город']=hh_data['Город, переезд, командировки'].apply(city_)# создаем новый признак "Город"
hh_data['Город'].nunique()#проверяю работу функции, только 4 категории должно быть
print(round(hh_data['Город'].value_counts(normalize=True)*100))# находим  долю городов
print("Процент соискателей живущих в Санкт-Петербурге:",round(hh_data['Город'].value_counts(normalize = True)['Санкт-Петербург']*100))
```

```python
hh_data['Город, переезд, командировки'].unique()## смотрим как представлена информация
```

```python
#функция преобразования готовности к переезду
def moving(arg):
    if ('не готов к переезду' in arg) or ('не готова к переезду' in arg):
        return False
    else:
        return True
hh_data['Готовность к переезду']=hh_data['Город, переезд, командировки'].apply(moving)# новый признак 

```

```python
# функция преобразования готовности к командировкам
def business_trip(arg):
    if ('командировка' in arg):
        if ('не готов к командировкам' in arg) or ('не готова к командировкам' in arg):
            return False
        else:
            return True
    else:
        return False    
hh_data['Готовность к командировкам']=hh_data['Город, переезд, командировки'].apply(business_trip)#новый признак
hh_data['Готовность к командировкам'].nunique()#проверка
```

```python
res=round((hh_data[(hh_data['Готовность к переезду']==True) & (hh_data['Готовность к командировкам']==True)].shape[0])/hh_data.shape[0]*100)
print('Процент одновременно готовых к командировкам и переезду равен:',res)
```

```python

def selary_s(arg):#функция для выделения суммы 
    selary_sum = float(arg.split(' ')[0])
    return selary_sum
def selary_c(arg):# функция для выделения вылюты
    selary_carr=arg.split(' ')[1]
    return selary_carr
hh_data_drop_1['сумма зп']=hh_data_drop_1['ЗП'].apply(selary_s)   
hh_data_drop_1['валюта зп']=hh_data_drop_1['ЗП'].apply(selary_c)

# создали словарь для перевода валюты
currency ={'руб.': 'RUB', 'KZT': 'KZT', 'USD': 'USD', 'бел.руб.': 'BYN', 'EUR': 'EUR', 'грн.': 'UAH', 'сум': 'UZS', 'KGS': 'KGS', 'AZN': 'AZN'} 
hh_data_drop_1['валюта зп']=hh_data_drop_1['валюта зп'].replace(currency)# привели валюту в зп к стандарту 
hh_data_drop_1['валюта зп'].unique()# проверили уникальные значения, все ли заменили
```

5.6. Количество выбросов методом z-отклонений. 

```python
#функция алгоритма метода z-отклонения
def outliers_z_score(data,feature,log_scale=False):
    if log_scale:
        x=np.log(data[feature]+1)
    else:
        x=data[feature]
    mu=x.mean()
    sigma=x.std()
    lower_bound=mu-3*sigma
    upper_bound=mu+4*sigma
    outliers=data[(x<lower_bound)|(x>=upper_bound)]
    cleaned=data[(x>=lower_bound)&(x<=upper_bound)]
    return outliers, cleaned
outliers, cleaned=outliers_z_score(hh_data_drop_4, 'Возраст', log_scale=True)
print(f'Число выбросов по методу Z -отклонения: {outliers.shape[0]}')
Z_age=outliers['Возраст'].unique()
#print(type(Z_age))
Z_age=str(Z_age)
#print(type(Z_age))
Z_age=Z_age.split()
Z_age=','.join(Z_age)

#print(Z_age)
print(f'В выбросы попадают соискатели с возрастом:{Z_age} лет')
```
